### DATA 준비

```python
import pandas as pd
import numpy as np
import os
import re
import sys
```

```python
#Review data
Review_fin = pd.read_csv('data/project_review.csv', encoding='cp949', engine = 'python')
Review_fin.shape
```

```python
#pos_tag 

def tagged_to_list(a):
    a = a.replace('\'','')
    a = a.replace('[','')
    a = a.replace(']','')
    return a

pos_list = [tagged_to_list(a) for a in Review_fin['pos_tagged']]
pos_list_s = [b.replace(',','').split(' ') for b in pos_list]
```

```python
Review_fin_re = pd.concat([Review_fin.drop(['reple','pos_tagged'], axis = 1), 
                       pd.DataFrame({'pos_list' : pos_list_s})], axis= 1, ignore_index=True)
Review_fin_re.columns = ['bad','good','score','id','y','pos_tag']
Review_re=pd.concat([Review_fin_re[Review_fin_re['score'] >= 8], 
                     Review_fin_re[Review_fin_re['score'] <= 3]], axis = 0, ignore_index=True)
```

```python
def preprocessing(line):
    
    text_temp = []
    
    for word in line:
        word_ = word.split('/')[0]
        pumsa_ = word.split('/')[1]
        
        if pumsa_.find('Noun' or 'Verb' or 'Adjective' \
                        or 'Foreign' or 'KoreanParticle' or 'Punctuation') == 0 \
                        and len(word_) >= 2:
            
            text_temp.append(word_)
        
    return(text_temp)  #unlist ' '.join()
```

```python
# Review_re에 빈 행 제거 
proc_list = list()
err_idx = list()
for i, pos_tag in enumerate(Review_re['pos_tag']):
    #print(i)
    try:
        proc_list.append(preprocessing(pos_tag))
    except IndexError as e:
        err_idx.append(i)
        
# 빈 리스트 찾기 
Review_re['pos_tag'][err_idx]

# err_idx remove
Review_re = Review_re.drop(err_idx)
```

```python
# preprocessing 처리
Review_tag = Review_re['pos_tag'].apply(lambda x : preprocessing(x))
```

```python
# 전처리 후에 빈 리스트 아이디 뽑자
null_idx = [idx for idx,tag in enumerate(Review_tag) if len(tag)==0]

for elem in err_idx:
    if elem in null_idx:
        print(elem)
```

```python
Review_tag = Review_tag.drop(null_idx)
real_idx = Review_tag.index
```

```python
# Data reproc
X = pd.concat([Review_re.drop('pos_tag', axis=1), Review_tag], axis = 1)
Y = Review_re['y']
```

```python
#t rain test split
from sklearn.cross_validation import train_test_split # 버전에 따라 다름

x_train, x_test, y_train, y_test = train_test_split(X.drop(null_idx), Y.drop(null_idx),
                                             test_size = 0.1, random_state = 0)
```

# Word2vec to Lasso

-  위의 X 라는 DATA로 Lasso 계수 뽑기
-  위의 x_train이라는 DATA로 Lasso 계수 뽑기
-  Lasso 계수 뽑기 전에 Word2vec min_count 다르게 해서 전체 단어 갯수 다르게 하기 

### Example : Data가 커서 x_test 를 이용해서 돌려보겠다.

```python
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
warnings.filterwarnings(action='ignore', category=UserWarning, module='gensim')
import gensim

from sklearn.feature_extraction.text import TfidfVectorizer
from gensim.models import Word2Vec
from collections import defaultdict


test_tag = x_test['pos_tag']
Review_word = list(test_tag)

# 20개 이하인 단어는 제거 
model = Word2Vec(Review_word, size = 100, window = 10, min_count = 20, workers = 4)
w2v = dict(zip(model.wv.index2word, model.wv.syn0))
```

```python
# min_count 조절하면서 단어 갯수 바꿔보자
len(model.wv.index2word) #1249 개 data
```

```python
# w2v_token : w2v 했을 때 사용하는 단어
# 아래의 lasso 에 사용하려고 따로 저장 

w2v_token = model.wv.index2word 
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer

# w2v의 model과 test data의 pos_tag부분만 넣어준다.
def w2v_tfidf(model, test_tag):
    
    w2v_token = model.wv.index2word  
    
    #countvect
    w2v_vect = CountVectorizer(vocabulary=w2v_token)
    w2v_tf = w2v_vect.fit_transform(test_tag.apply(lambda x : ' '.join(x))) #unlist
    
    #tfidf
    w2v_tfidf = TfidfTransformer(norm = 'l2')
    w2v_tfidf.fit(w2v_tf)
    
    w2v_tfidf_mat = w2v_tfidf.transform(w2v_tf)
    w2v_tfidf_df = pd.DataFrame(w2v_tfidf_mat.todense(), columns = model.wv.index2word)
    
    return w2v_tfidf_df
```

```python
w2v_tfidf_df = w2v_tfidf(model, test_tag)
```

```python
w2v_tfidf_df
```

### 드디어 Lasso로 변수 선택!
- Data가 커지면 Memory Error 
- Desktop으로 돌리기

```python
from sklearn.linear_model import RidgeCV, LassoCV, ElasticNetCV

lasso = LassoCV(alphas = [ 0.002,0.004,0.005, 0.007,
                          0.0001, 0.0003, 0.0006, 0.001, 0.00003, 0.00006, 1,2,3,4])

# 위의 w2v_tfidf와 x_test에 대응되는 y_test를 넣어준다.
lasso.fit(w2v_tfidf_df, y_test)
alpha = lasso.alpha_

print("Best alpha : ", alpha)
```

```python
# parameter tuning
lasso = LassoCV(alphas = [alpha *.6, alpha * .65, alpha * .7, alpha * .75,
                         alpha * .8, alpha * .85, alpha * .9, alpha * .95,
                         alpha * 1.05, alpha * 1.1, alpha * 1.15, alpha * 1.25,
                         alpha * 1.3, alpha * 1.35, alpha * 1.4],
               
                max_iter = 50000, cv=10)

lasso.fit(w2v_tfidf_df, y_test)
alpha = lasso.alpha_
print("Best alpha : ", alpha)


coefs = pd.Series(lasso.coef_, index = w2v_token)
print("lasso는 변수를 몇개 선택했니? : ", sum(coefs!=0)) 
print("lasso는 변수를 몇개 버렸니? : ",sum(coefs==0)) 
```

```python
las_coef_df = pd.DataFrame({'value' : lasso.coef_, 'word' : w2v_token})

#las_coef에서 0인 값 제거
w2v_las_df= las_coef_df[las_coef_df['value'] != 0]

#reset_index
w2v_las_df = w2v_las_df.reset_index(drop='index')
w2v_las_df
```
